Your task in this week’s assignment is to answer three questions: 
1. What is the northernmost airport in the United States? 
2. What is the easternmost airport in the United States? 
3. On February 12th, 2013, which New York area airport had the windiest weather?

Use the information in the .CSV files found at https://github.com/hadley/nycflights13/tree/master/data-raw for your source data.

### 1. Nothernmost Airport
The CSV file located at https://github.com/tidyverse/nycflights13/blob/master/data-raw/airports.csv contains airport location information, by latitude and longitude. The northernmost locations will have higher degree values of latitude. Pandas can be used to sort the data to find the highest degree of latitude in the dataset.

##### First, the pandas module is loaded, followed by the raw CSV file:

```python
import pandas as pd
pd.set_option('display.max_rows', 10)

aports = pd.read_csv('https://raw.githubusercontent.com/tidyverse/nycflights13/master/data-raw/airports.csv')
```

#### With the data loaded, pandas can be instructed to retrieve the highest Latitude value in the list:

```python
aports.sort_values(by = ['lat'], ascending = False)
```

#### The northermost airport, according to this data is *Dillant Hopkins Airport*.

### 2. Easternmost Airport
A similar approach can be taken to find the easternmost airport, using the same dataset. This time, however, the logitude values will be used:

```python
aports.sort_values(by = ['lon'], ascending = False)
```

#### The easternmost airport, according to this data is *Eareckson Air Station*

### 3. Windiest Weather
The dataset located at https://github.com/tidyverse/nycflights13/blob/master/data-raw/weather.csv contains weather
information for the airports. Like the previous questions, pandas can be used to sort the data to find the answer.

First, the CSV file is loaded:

```python
weather = pd.read_csv('https://raw.githubusercontent.com/tidyverse/nycflights13/master/data-raw/weather.csv')
```

Now the data can be sorted by the values that will be needed, such as the *day/month/year* and *wind speed*, using pandas:

```python
weather[(weather["year"] == 2013) & (weather["month"] == 2) & (weather["day"] == 12)].sort_values(by = ["wind_speed"], ascending = False)
```

#### According to the data, Newark Liberty International Airport had the windiest weather on February 12th, 2013.
